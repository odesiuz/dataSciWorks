```python
from keras.applications.resnet50 import ResNet50
from keras.preprocessing import image
from keras.applications.resnet50 import preprocess_input, decode_predictions
import numpy as np

model = ResNet50(weights='imagenet') # we specify that we want to obtain the weight based on a training which has been done on the imagenet data set.

img_path = 'lion.jpg'
img = image.load_img(img_path, target_size=(224, 224)) # We scale the image down to 224 by 224
x = image.img_to_array(img) # we turn this into a 3D-array known as a 3D tensor with 224 by 224 by 3. 3 signifies a coloured image
x.shape
```

```python
x = np.expand_dims(x, axis=0) # we now expand the shape to 1 by 224 by 224 by 3. 1 is because we are testing just one image
x = preprocess_input(x) # we normalise the data to range between -1 and +1
```

```python
x.shape
```

```python
preds = model.predict(x) # Now we predict the elephant picture. The output is a vector of 5000. So we get back a vector with class probabilities 100 which is not very handy
```

```python
preds
```

```python
# decode the results into a list of tuples (class, description, probability)
# (one such list for each sample in the batch)
print('Predicted:', decode_predictions(preds, top=3)[0]) # We use the decode_prediction methods to handle the 1000 class vectors and hence we get the actual class labels as strings. top = 3 signifies the top 3 best classes
# Predicted: [(u'n02504013', u'Indian_elephant', 0.82658225), (u'n01871265', u'tusker', 0.1122357), (u'n02504458', u'African_elephant', 0.061040461)]
```
